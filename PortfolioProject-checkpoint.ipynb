```python
import seaborn as sns
```

```python
mpg=sns.load_dataset("mpg")
mpg
```

## 1) Create a visual showing the relationship between horsepower and fuel economy.

```python
sns.relplot(data=mpg, x="mpg", y="horsepower", hue= "weight", 
        )
```

## 1) Comments:

In this scatterplot, we can see that the fuel economy is not driven by horsepower alone as weight also plays a major role in how horsepower is translated into mpg. This implies that improving fuel economy is most effective through weight reduction and imporovements in efficiency, not just weight alone. 

## 2) Create two other visuals or tables related to fuel economy.


## Visual 1, 

```python
sns.relplot(
    data=mpg,
    x="mpg", y="displacement",
    hue="cylinders", palette="ch:r=-.5,l=.75"
)
```

## 2a) Comments 

In this graph, we are shown the relationship between fuel economy and engine displacement. where the larger engines generally achieve lower mpg. Colouring by cyilinder count reveals clear grouping, with higher cylinder engines clustering at larger displacement and poorer fuel effciency compared to lower cylinder vehicles.

## Visual 2 

```python
sns.barplot(data=mpg, x = "origin", y= "mpg", width = 0.5, errorbar=None)
```

## 2b) Comment 

This graph shows us how different countries has varying levels of fuel efficency. We can see base on the chart that the country with the best fuel economy is going to be japan. 

## 3) Create a visual showing how fuel economy has changed over time


```python
sns.relplot(data=mpg, x="model_year", y="mpg",  kind= "line")
```

## 3) Comment

In this line visual here, we can see how fuel economy has changed over time, with the climb for improvement starting in 75' and peaking in the 80's.
